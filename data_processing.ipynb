```python
import os
import cv2
import numpy as np
from tqdm import tqdm
import matplotlib.pyplot as plt
```

```python
import os

violence_path = r"dataset\Real Life Violence Dataset\Violence"
nonviolence_path = r"dataset\Real Life Violence Dataset\NonViolence"

print("Violence videos:", len(os.listdir(violence_path)))
print("NonViolence videos:", len(os.listdir(nonviolence_path)))
```

```python
processed_path = r"processed_data"

violence_processed = os.path.join(processed_path, "V")
nonviolence_processed = os.path.join(processed_path, "NV")

os.makedirs(violence_processed, exist_ok=True)
os.makedirs(nonviolence_processed, exist_ok=True)

print("Processed folders ready")
```

```python
import cv2
import os

sample_folder = violence_path   # check violence videos first

videos = os.listdir(sample_folder)[:5]  # check first 5 videos

for video in videos:
    
    video_path = os.path.join(sample_folder, video)
    
    cap = cv2.VideoCapture(video_path)
    
    fps = cap.get(cv2.CAP_PROP_FPS)
    frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    
    width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    
    duration = frame_count / fps if fps != 0 else 0
    
    print("Video:", video)
    print("FPS:", fps)
    print("Frames:", frame_count)
    print("Resolution:", width, "x", height)
    print("Duration:", round(duration,2), "seconds")
    print("------------")
    
    cap.release()
```

```python
def extract_frames(video_path, save_dir, label, frames_per_video=16, img_size=160):

    cap = cv2.VideoCapture(video_path)

    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

    if total_frames < frames_per_video:
        cap.release()
        return

    frame_indices = np.linspace(0, total_frames - 1, frames_per_video).astype(int)

    frame_id = 0
    saved = 0

    while True:

        ret, frame = cap.read()

        if not ret:
            break

        if frame_id in frame_indices:

            frame = cv2.resize(frame, (img_size, img_size))

            filename = f"{label}_{os.path.basename(video_path).split('.')[0]}_{saved}.jpg"

            cv2.imwrite(os.path.join(save_dir, filename), frame)

            saved += 1

        frame_id += 1

    cap.release()
```

```python
for video in tqdm(os.listdir(violence_path)):

    video_path = os.path.join(violence_path, video)

    extract_frames(video_path, violence_processed, label="V")
```

```python
for video in tqdm(os.listdir(nonviolence_path)):

    video_path = os.path.join(nonviolence_path, video)

    extract_frames(video_path, nonviolence_processed, label="NV")
```
